```python
#For each tweet
#For each word in tweet
#Get the word score (score is a numberif the word is in Lexicon, 0 if not)
#Add all the scores and find the ploarity
strength = []
prediction = []
for tweet in data_list:
    score = 0
    for word in tweet:
        if word in (lexicons):
            score = score + lexicons[word]
    strength.append(score)
    if (score >= 0):
        prediction.append('ham')
    else:
        prediction.append('spam')
print(strength)
print(prediction)
```
